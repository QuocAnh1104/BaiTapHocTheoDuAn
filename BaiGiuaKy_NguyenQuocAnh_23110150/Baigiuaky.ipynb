```python
def bai1_todo_list_interactive():
    print("--- ỨNG DỤNG TO-DO LIST ---")
    task_list = []
    
    # Cho phép người dùng nhập các công việc
    while True:
        task = input("Nhập công việc (hoặc gõ 'xong' để dừng): ")
        
        # Thêm .strip() để xóa khoảng trắng thừa, .lower() để đưa về chữ thường
        if task.strip().lower() == 'xong':
            break
            
        # Kiểm tra thêm: nếu người dùng không nhập gì mà bấm Enter luôn thì bỏ qua
        if task.strip() != "":
            task_list.append(task)
    
    if not task_list:
        print("Danh sách trống.")
        return

    print("\nDanh sách công việc của bạn:")
    numbered_tasks = []
    for i, task in enumerate(task_list, 1):
        line = f"{i}. {task}"
        print(line)
        numbered_tasks.append(line)
        
    try:
        with open("todo_list.txt", "w", encoding="utf-8") as f:
            for item in numbered_tasks:
                f.write(item + "\n")
        print("\n-> Đã lưu thành công vào file 'todo_list.txt'.")
    except Exception as e:
        print(f"Có lỗi khi lưu file: {e}")

if __name__ == "__main__":
    bai1_todo_list_interactive()
```

```python
import re
from collections import Counter

def bai2_phan_tich_interactive():
    print("--- PHÂN TÍCH VÀ THỐNG KÊ VĂN BẢN ---")
    
    # Yêu cầu người dùng nhập văn bản
    text = input("Nhập vào một đoạn văn bản (khoảng 2-3 câu): ").strip()
    
    # Kiểm tra nếu người dùng chỉ nhấn Enter mà không nhập gì
    if not text:
        print("Bạn chưa nhập văn bản nào! Vui lòng chạy lại chương trình.")
        return
        
    print(f"\nĐoạn văn bạn vừa nhập: '{text}'")
    
    # Chuyển thành chữ thường và loại bỏ dấu câu cơ bản (chấm, phẩy)
    # Bạn có thể thêm các dấu câu khác vào trong r'[.,!?]' nếu cần
    cleaned_text = re.sub(r'[.,]', '', text).lower()
    words = cleaned_text.split()
    
    # In ra danh sách các từ đã được tách
    print(f"-> Danh sách các từ trong văn bản: {words}")
    
    # Đếm tổng số từ
    total_words = len(words)
    print(f"-> Tổng số từ trong đoạn văn: {total_words}")
    
    # Tìm TẤT CẢ các từ xuất hiện nhiều nhất
    if total_words > 0:
        # Đếm tần suất xuất hiện của từng từ
        word_counts = Counter(words)
        
        # Tìm số lần xuất hiện cao nhất
        max_count = max(word_counts.values())
        
        # Lọc ra danh sách các từ có số lần xuất hiện bằng với max_count
        most_common_words = [word for word, count in word_counts.items() if count == max_count]
        
        # In ra kết quả tùy thuộc vào việc có 1 hay nhiều từ
        if len(most_common_words) == 1:
            print(f"-> Từ xuất hiện nhiều nhất là: '{most_common_words[0]}' (xuất hiện {max_count} lần)")
        else:
            # Nối các từ lại với nhau cho đẹp mắt
            words_str = "', '".join(most_common_words)
            print(f"-> Các từ xuất hiện nhiều nhất là: '{words_str}' (cùng xuất hiện {max_count} lần)")

# Chạy chương trình
if __name__ == "__main__":
    bai2_phan_tich_interactive()
```

```python
def bai3_bmi_interactive():
    print("--- CÔNG CỤ TÍNH CHỈ SỐ BMI THÔNG MINH ---")
    print("(Gõ 'thoát' ở ô nhập dữ liệu nếu bạn muốn dừng chương trình)\n")
    
    while True:
        # 1. Nhập cân nặng
        weight_input = input("Nhập cân nặng của bạn (kg): ").strip()
        if weight_input.lower() == 'thoát':
            print("-> Đã thoát chương trình tính BMI.")
            break
            
        # 2. Nhập chiều cao
        height_input = input("Nhập chiều cao của bạn (m, ví dụ: 1.75): ").strip()
        if height_input.lower() == 'thoát':
            print("-> Đã thoát chương trình tính BMI.")
            break
        
        # 3. Xử lý tính toán và bắt lỗi bằng try-except theo yêu cầu đề bài
        try:
            # Ép kiểu dữ liệu sang số thực (float)
            weight = float(weight_input)
            height = float(height_input)
            
            # Tính BMI
            bmi = weight / (height ** 2)
            print(f"\n=> Chỉ số BMI của bạn là: {bmi:.2f}")
            
            # Phân loại kết quả
            if bmi < 18.5:
                print("=> Phân loại: Gầy\n")
            elif 18.5 <= bmi < 25: # Bao phủ khoảng từ 18.5 đến dưới 25 (Bình thường)
                print("=> Phân loại: Bình thường\n")
            else:
                print("=> Phân loại: Thừa cân\n")
                
        except ValueError:
            # Lỗi khi người dùng nhập chữ (ví dụ: "sáu mươi", "một mét bảy")
            print("\n[LỖI] Bạn đã nhập chữ hoặc sai định dạng số! Vui lòng chỉ nhập số (ví dụ: 65 hoặc 1.75).\n")
            
        except ZeroDivisionError:
            # Lỗi khi người dùng nhập chiều cao = 0
            print("\n[LỖI] ZeroDivisionError: Chiều cao không thể bằng 0! Vui lòng nhập lại.\n")

# Chạy chương trình
if __name__ == "__main__":
    bai3_bmi_interactive()
```

```python
import random
import time

def bai4_hangman_interactive():
    print("--- TRÒ CHƠI ĐOÁN CHỮ CÁI (MINI HANGMAN) ---")
    word_list = ["python", "apple", "river", "ocean", "mountain"]
    secret_word = random.choice(word_list)
    
    guessed_letters = set()
    wrong_turns = 5
    start_time = time.time()
    
    # Thêm biến kiểm tra trạng thái thắng/thua
    is_win = False 
    
    print("Máy tính đã chọn một từ tiếng Anh bí mật!")
    
    while wrong_turns > 0:
        current_display = ""
        for char in secret_word:
            if char in guessed_letters:
                current_display += char + " "
            else:
                current_display += "_ "
        
        if "_" not in current_display:
            print(f"\n=> CHÚC MỪNG! Bạn đã chiến thắng. Từ bí mật là '{secret_word}'")
            is_win = True  # Đánh dấu là đã thắng
            break
            
        print(f"\nTừ hiện tại: {current_display}")
        print(f"Lượt sai cho phép còn lại: {wrong_turns}")
        
        guess = input("Nhập một chữ cái để đoán (hoặc gõ 'thoat' để dừng): ").strip().lower()
        
        if guess == 'thoat':
            print("Bạn đã thoát trò chơi nửa chừng.")
            return
            
        if len(guess) != 1 or not guess.isalpha():
            print("-> [Lỗi] Vui lòng chỉ nhập MỘT chữ cái từ A-Z!")
            continue
            
        if guess in guessed_letters:
            print(f"-> Bạn đã đoán chữ '{guess}' trước đó rồi. Hãy thử chữ khác!")
            continue
            
        guessed_letters.add(guess)
        
        if guess not in secret_word:
            wrong_turns -= 1
            print(f"-> Sai rồi! Chữ '{guess}' KHÔNG CÓ trong từ bí mật.")
        else:
            print(f"-> Chính xác! Chữ '{guess}' CÓ trong từ bí mật.")

    if not is_win:
        print(f"\n=> RẤT TIẾC, BẠN ĐÃ THUA! Từ bí mật chính là '{secret_word}'")

    end_time = time.time()
    time_taken = round(end_time - start_time, 2)

    # Đưa trạng thái Thắng/Thua vào file text
    status_text = "Thắng" if is_win else "Thua"

    try:
        with open("hangman_record.txt", "a", encoding="utf-8") as f:
            f.write(f"Kết quả: {status_text} | Từ: {secret_word} | Lượt sai còn lại: {wrong_turns} | Thời gian: {time_taken} giây\n")
        print(f"\n-> Đã lưu lịch sử chơi (Trạng thái: {status_text}) vào file 'hangman_record.txt'.")
    except Exception as e:
        print(f"Có lỗi khi lưu file: {e}")

if __name__ == "__main__":
    bai4_hangman_interactive()
```

```python
import matplotlib.pyplot as plt

def bai5_kho_hang_to_uu():
    # Dữ liệu tồn kho ban đầu
    kho = {'Balo': 150, 'Tui xach': 80, 'Vali': 120}
    
    print("--- ỨNG DỤNG QUẢN LÝ KHO HÀNG ---")
    
    while True:
        print("\n" + "="*45)
        print("1. Xem danh sách tồn kho")
        print("2. Thêm mặt hàng mới")
        print("3. Cập nhật số lượng mặt hàng")
        print("4. Vẽ, xem và lưu biểu đồ tỷ trọng (Pie Chart)")
        print("5. Thoát")
        print("="*45)
        
        chon = input("Nhập lựa chọn của bạn (1-5): ").strip()
        
        if chon == '1':
            print("\n[Tồn kho hiện tại]")
            for item, qty in kho.items():
                print(f" - {item}: {qty}")
                
        elif chon == '2':
            print("\n[Thêm mặt hàng mới]")
            ten_hang = input("Nhập tên mặt hàng mới (VD: Vi da): ").strip()
            if ten_hang in kho:
                print(f"-> [Lỗi] Mặt hàng '{ten_hang}' đã có trong kho! Hãy dùng chức năng (3) để cập nhật.")
            else:
                try:
                    # Nhập luôn số lượng ngay sau khi nhập tên
                    so_luong = int(input(f"Nhập số lượng ban đầu cho '{ten_hang}': ").strip())
                    if so_luong < 0:
                        print("-> [Lỗi] Số lượng không được âm!")
                    else:
                        kho[ten_hang] = so_luong
                        print(f"-> Đã thêm thành công '{ten_hang}' với số lượng: {so_luong}")
                except ValueError:
                    print("-> [Lỗi] Vui lòng nhập số nguyên hợp lệ!")
                    
        elif chon == '3':
            print("\n[Cập nhật số lượng mặt hàng]")
            ten_hang = input("Nhập tên mặt hàng cần cập nhật: ").strip()
            if ten_hang not in kho:
                print(f"-> [Lỗi] Không tìm thấy '{ten_hang}' trong kho.")
            else:
                try:
                    # Nhập thẳng số lượng mới thay vì tính toán cộng trừ
                    so_luong_moi = int(input(f"Nhập SỐ LƯỢNG MỚI cho '{ten_hang}': ").strip())
                    if so_luong_moi < 0:
                        print("-> [Lỗi] Số lượng không được âm!")
                    else:
                        kho[ten_hang] = so_luong_moi
                        print(f"-> Đã cập nhật! Số lượng '{ten_hang}' hiện tại là: {kho[ten_hang]}")
                except ValueError:
                    print("-> [Lỗi] Vui lòng nhập số nguyên hợp lệ!")
                    
        elif chon == '4':
            print("\n[Vẽ biểu đồ]")
            if sum(kho.values()) == 0:
                print("-> Kho hàng đang trống hoặc tổng số lượng = 0, không thể vẽ biểu đồ.")
                continue
                
            labels = list(kho.keys())
            sizes = list(kho.values())
            
            plt.figure(figsize=(7, 7))
            plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, 
                    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0'])
            plt.title('Tỷ Trọng Hàng Hóa Trong Kho')
            
            file_name = "bieu_do_kho_hang.png"
            plt.savefig(file_name)
            print(f"-> Đã lưu biểu đồ vào file '{file_name}'.")
            
            print("-> Đang hiển thị biểu đồ... (Vui lòng tắt cửa sổ biểu đồ để tiếp tục)")
            plt.show()
            
        elif chon == '5':
            print("\n-> Đã thoát chương trình. Chúc bạn một ngày tốt lành!")
            break
            
        else:
            print("\n-> [Lỗi] Lựa chọn không hợp lệ. Vui lòng nhập từ 1 đến 5.")

if __name__ == "__main__":
    bai5_kho_hang_to_uu()
```
