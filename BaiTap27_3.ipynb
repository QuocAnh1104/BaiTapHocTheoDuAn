```python
from unittest.mock import patch, MagicMock
from datetime import datetime
import requests
```

```python
# =========================
# Bài 1. Hàm tính thuế
# =========================

def calculate_tax(income):
    if income < 5000:
        return 0
    elif income < 10000:
        return income * 0.1
    else:
        return income * 0.2


def test_calculate_tax():
    assert calculate_tax(3000) == 0        # income < 5000
    assert calculate_tax(7000) == 700      # income = 7000
    assert calculate_tax(12000) == 2400    # income = 12000
```

```python
# =========================
# Bài 2. Kiểm tra số nguyên tố
# =========================

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


def test_is_prime():
    assert is_prime(1) == False
    assert is_prime(2) == True
    assert is_prime(3) == True
    assert is_prime(4) == False
    assert is_prime(17) == True
    assert is_prime(18) == False
    assert is_prime(19) == True
```

```python
# =========================
# Bài 3. Test lớp User
# =========================

class User:
    def __init__(self, username):
        self.username = username

    def is_admin(self):
        return self.username == "admin"


def test_user_is_admin():
    assert User("admin").is_admin() == True
    assert User("guest").is_admin() == False
    assert User("user123").is_admin() == False
```

```python
# =========================
# Bài 4. Mock gửi email
# =========================

def send_welcome_email(email):
    print(f"Sending email to {email}")


@patch("builtins.print")
def test_send_welcome_email(mock_print):
    send_welcome_email("test@gmail.com")
    mock_print.assert_called_once_with(
        "Sending email to test@gmail.com"
    )
```

```python
# =========================
# Bài 5. Mock API requests.get
# =========================

def fetch_user():
    response = requests.get("https://api.example.com/user")
    return response.json()


@patch("requests.get")
def test_fetch_user(mock_get):
    mock_response = MagicMock()
    mock_response.json.return_value = {
        "id": 1,
        "name": "John"
    }

    mock_get.return_value = mock_response

    result = fetch_user()

    assert result == {
        "id": 1,
        "name": "John"
    }

    mock_get.assert_called_once_with(
        "https://api.example.com/user"
    )
```

```python
# =========================
# Bài 6. Mock ngày hiện tại
# =========================

def is_weekend():
    today = datetime.now().weekday()
    return today >= 5


@patch("datetime.datetime")
def test_is_weekend_friday(mock_datetime):
    mock_datetime.now.return_value = datetime(2026, 4, 24)  # Friday
    # weekday = 4
    assert is_weekend() == False


@patch("datetime.datetime")
def test_is_weekend_saturday(mock_datetime):
    mock_datetime.now.return_value = datetime(2026, 4, 25)  # Saturday
    # weekday = 5
    assert is_weekend() == True
```

```python
# =========================
# Bài 7. Test mật khẩu mạnh
# =========================

def is_strong(password):
    return len(password) >= 8 and any(c.isdigit() for c in password)


def test_is_strong():
    assert is_strong("abc") == False
    assert is_strong("abcdefgh") == False
    assert is_strong("abc123") == False
    assert is_strong("password1") == True
    assert is_strong("StrongPass99") == True
```

```python
# =========================
# Bài 8. Test xử lý chuỗi
# =========================

def clean_input(s):
    return s.strip().lower().replace(" ", "_")


def test_clean_input():
    assert clean_input(" Hello World ") == "hello_world"
    assert clean_input("PYTHON TEST") == "python_test"
    assert clean_input("   many   spaces   ") == "many___spaces"
```

```python
# =========================
# Bài 9. Test chia số
# =========================

def safe_divide(a, b):
    if b == 0:
        return None
    return a / b


def test_safe_divide():
    assert safe_divide(10, 2) == 5
    assert safe_divide(5, 0) == None
    assert safe_divide(-4, 2) == -2
```
